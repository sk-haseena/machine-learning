```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("../weather.csv")[:100]

features = np.array(dataset["Humidity"])
labels = np.array(dataset['Temperature (C)'])

print(dataset.columns)
```

```python
w = np.random.randn()
b = np.random.randn()

print(w,b)
```

```python
preds = w*features + b

plt.scatter(features, labels)
plt.plot(features, preds, color="purple")
plt.show()
```

```python
def mse(preds, labels):
    return np.sum((preds - labels)**2)/len(labels) # mean squared error
loss = mse(preds, labels)
print(loss)
```

```python
# gradient descent
# -(y - y')* x <---- w
# -(y - y')    <---- b

w_grad = np.sum(-1*(labels - preds) * features)/len(labels)
b_grad = np.sum(-1*(labels - preds))/len(labels)

print(w_grad, b_grad)
```

```python
lr = 0.1

if w_grad < 0:
    w += lr
else:
    w -= lr

if b_grad < 0:
    b += lr
else:
    b -= lr
```

```python
print(w, b)
```

```python
for i in range(1000):
    w_grad = np.sum(-1*(labels - preds) * features)/len(labels)
    b_grad = np.sum(-1*(labels - preds))/len(labels)

    lr = 0.0001

    if w_grad < 0:
        w += lr
    else:
        w -= lr

    if b_grad < 0:
        b += lr
    else:
        b -= lr

    preds = w*features + b

    loss = mse(preds, labels)
    print(loss)
```
